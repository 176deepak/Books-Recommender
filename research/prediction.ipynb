```python
import numpy as np

def recommender(book_name):
    index = np.where(loaded_model.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(loaded_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    books = []
    for i in similar_items:
        item = []
        temp_df = loaded_data[loaded_data['Book-Title'] == loaded_model.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        
        books.append(item)
    
    return books
```
